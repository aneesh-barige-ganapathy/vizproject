```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r"C:\Users\bgane\Desktop\vis project 1_files\cleaned_data.csv")

# Step 1: Check for missing values
# Display the count of missing values in each column
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)

# Step 2: Handle missing values
# For this dataset, we will:
# - Drop rows with missing values in any column to ensure data integrity

data.dropna(inplace=True)

# Step 3: Remove duplicates
# Check for duplicates in the dataset
duplicates = data.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")

# Drop duplicate rows
data = data.drop_duplicates()

# Step 4: Standardize column names
# Rename columns to make them more Python-friendly
data.rename(columns=lambda x: x.strip().replace(" ", "_"), inplace=True)



# Step 6: Convert emissions columns to numeric (if not already)
for year in range(2000, 2022):
    data[str(year)] = pd.to_numeric(data[str(year)], errors='coerce')

# Step 7: Save the cleaned data for use in visualizations
cleaned_file_path = r"C:\Users\bgane\Desktop\vis project 1_files\cleaned_file_path.csv"
data.to_csv(cleaned_file_path, index=False)

print(f"Cleaned data saved to {cleaned_file_path}")

# Display a summary of the cleaned data
print(data.info())
print(data.head())


#seperation of  total from sectors
filtered_data = data[data['sector'] == 'Total including LUCF']

file_path = r"C:\Users\bgane\Desktop\vis project 1_files\Total including LUCF.csv"
filtered_data.to_csv(file_path, index=False)

iltered_data = data[data['sector'] != 'Total including LUCF']

cleaned_file_path = r"C:\Users\bgane\Desktop\vis project 1_files\OTHER_SECTOR.csv"
iltered_data.to_csv(cleaned_file_path, index=False)
```
